```python
train_data = pd.read_csv("/kaggle/input/titanic-competition/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("/kaggle/input/titanic-competition/test.csv")
test_data.head()
```

```python
import pandas as pd 

combine = [train_data , test_data]

for df in combine : 

    df['Age'] = df['Age'].fillna(df['Age'].median())

    df['fare'] = df['Fare'].fillna(df['Fare'].median())

    df['Embarked'] = df['Embarked'].fillna('S')

    df['Sex'] = df['Sex'].map({'female': 1, 'male': 0}).astype(int)

    df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2}).astype(int)

    df['FamilySize'] = df['SibSp'] + df['Parch'] + 1 


# Define the features we will use for training

features = ["Pclass", "Sex", "Age", "Fare", "Embarked", "FamilySize"]

X = train_data[features]
y = train_data["Survived"]
X_test = test_data[features]

print("Data cleaning and feature engineering complete!")
X.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

# 1. Instantiate the model

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)

# 2. Fit (Train) the model using training data
model.fit(X, y)

print("Model training is complete!")
```

```python
# 3. Use the model to make predictions on the test set
predictions = model.predict(X_test)

print("Predictions generated for 418 passengers.")
```

```python
# 4. Prepare the final submission file
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})

# 5. Save as a CSV file in the Kaggle output directory
output.to_csv('submission.csv', index=False)

print("Your submission file 'submission.csv' has been saved!")
```
